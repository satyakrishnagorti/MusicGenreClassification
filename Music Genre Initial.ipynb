<h2>All Imports</h2>

```python
import os
import sys
from scipy.io import wavfile
import scipy
from pydub import AudioSegment
import matplotlib.pyplot as plt
import pylab
from matplotlib.pyplot import specgram
```

<h2>Script to Convert au to wav

```python
try:
    os.mkdir("data")
except:
    pass
for eachdir in os.listdir(INIT_DIR):
    try:
        os.mkdir(DATA_DIR+eachdir)
    except:
        pass
    for eachfile in os.listdir(INIT_DIR+eachdir):
        #print INIT_DIR+eachdir+"/"+eachfile
        #print eachfile
        song = AudioSegment.from_file(INIT_DIR+eachdir+"/"+eachfile,"au")
        song.export(DATA_DIR+eachdir+"/"+eachfile[:-2]+"wav",format="wav")
        
```

<h4>Testing specgram</h4>

```python
test_file = DATA_DIR+"blues/blues.00000.wav"
sample_rate, x = wavfile.read(test_file)
print sample_rate, x.shape
print type(sample_rate), type(x)
graph = specgram(x, Fs=sample_rate, xextent = (0,30))
pylab.show(graph)
```
